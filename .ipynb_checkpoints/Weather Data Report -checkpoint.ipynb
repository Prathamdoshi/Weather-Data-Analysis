```python
import pandas as pd
import openweathermapy.core as ow
import random 
import requests as req
```

```python
settings = {
    'appid': '84614c7c50570e5397bcdec7ce7bd427',
    'units': 'metric'
}
```

```python
cities = pd.read_csv('Cities/world_cities.csv')
cities = list(cities['name']) 
```

```python
random_cities = []

for i in range(0,500):
    random_number = random.randint(0,500)
    random_cities.append(cities[random_number])
```

```python
city_information = []

for city in random_cities:
   
    try: 
        response = ow.get_current(city,**settings)
        city_information.append(response)
        print('Found: {}'.format(city))
    except: 
        print("Couldn't Find: {}".format(city))
```
