```python
import pandas as pd
import openweathermapy.core as ow
import requests as req
import matplotlib.pyplot as plt
from citipy import citipy
import numpy as np
import time
```

```python
settings = {
    'appid': '84614c7c50570e5397bcdec7ce7bd427',
    'units': 'imperial'
}
```

```python
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lats_lngs = zip(lats,lngs)
```

```python
cities = []

for lat_lng in lats_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
cities 
```

```python
data_fetch = []

for city in cities:
    try: 
        response = ow.get_current(city, **settings) 
        cities_dict = {}
        cities_dict['City'] = city
        cities_dict['Cloudiness'] = response['clouds']['all']
        cities_dict['Humidity'] = response['main']['humidity']
        cities_dict['Max Temp'] = response['main']['temp_max']
        cities_dict['Wind Speed'] = response['wind']['speed']
        cities_dict['Long'] = response['coord']['lon']
        cities_dict['Lat'] = response['coord']['lat']
        data_fetch.append(cities_dict)        
        print("City No. {}, City Name: {}, Status: Success".format(cities.index(city),city))
        
    except: 
        print("ERROR: City No. {}, City Name: {}, Status: Fail".format(cities.index(city),city))
```

```python
cities_df = pd.DataFrame(data_fetch)
cities_df.head()
```

```python
plt.scatter(cities_df['Lat'],cities_df['Max Temp'],edgecolor="black", linewidths=1, marker="o", 
            alpha=0.8, label="Cities")

# Incorporate the other graph properties
plt.title("City Latitude vs. Max Temperature (%s)" % time.strftime("%x"))
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)
```
